```python
Heath=Heath.loc[Heath['miss']==True]
```

```python
Red_River_misses=Heath.loc[Heath['Trash Hauler']=='Red River']
```

```python
Waste_Ind_misses=Heath.loc[Heath['Trash Hauler']=='Waste Ind']
```

```python
Metro_misses=Heath.loc[Heath['Trash Hauler']=='Metro']
```

```python
Metro_multiple_misses=Metro_misses.groupby('Incident Address Street')['miss'].count()
Metro_multiple_misses=Metro_multiple_misses.reset_index()
```

```python
Waste_Ind_multiple_misses=Waste_Ind_misses.groupby('Incident Address Street')['miss'].count().reset_index()
```

```python
Waste_Ind_multiple_misses=Waste_Ind_multiple_misses.loc[Waste_Ind_multiple_misses['miss']>=2]
```

```python
Red_River_multiple_misses=Red_River_misses.groupby('Incident Address Street')['miss'].count().reset_index()
```

```python
Red_River_multiple_misses=Red_River_multiple_misses.loc[Red_River_multiple_misses['miss']>=2]
```

```python
Metro_multiple_misses=Metro_multiple_misses.loc[Metro_multiple_misses['miss']>=2]
```

```python
Metro_multiple_misses['Trash Hauler'] = 'Metro'
Waste_Ind_multiple_misses['Trash Hauler']= 'Waste_Ind'
Red_River_multiple_misses['Trash Hauler']= 'Red_River'
```

```python
concat_data_1 = pd.concat([Metro_multiple_misses, Waste_Ind_multiple_misses])
concat_data_1

## Notice that the index repeats itself
```

```python
Haulers_merged = pd.merge(concat_data_1, Red_River_multiple_misses, 
                               left_on = 'miss', right_on = 'miss', 
                               how = 'outer')
Haulers_merged.head(3)
Haulers_merged.tail(3)
```
